```python
# import librarys
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. sin graph
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.title('sin graph')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x, y, label = 'sin')
plt.legend()
```

```python
# 2. sin & cos graph in one plot.
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x, y1, label='sin')
plt.plot(x, y2, label='cos')
plt.title('sin & cos graph')
plt.legend()
```

```python
# 3. subplot
# subplot(row, col, which number plot?)

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.subplot(211)
plt.title('sin graph')
plt.plot(x, y1, label = 'sin')
plt.legend()
plt.subplot(212)
plt.title('cos graph')
plt.plot(x, y2, label = 'cos')
plt.tight_layout()
plt.legend()
```

```python
# 4. subplot use add_subplot
fig = plt.figure()
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
splot1 = fig.add_subplot(311)
splot1.plot(x, y1)
splot2 = fig.add_subplot(312)
splot2.plot(x, y2)
splot3 = fig.add_subplot(313)
splot3.plot(x, y3)
plt.tight_layout()
```

```python
# 5. scatter

n = 100
x = np.linspace(0, 100, n)
y = np.random.randn(n)
plt.scatter(x, y)
```

```python
# 6. hist
y = np.random.randn(1000)
plt.hist(y)
```

```python
# 7. DateFormatter 
from datetime import datetime, timedelta
import matplotlib.dates as mdate
days = 30
x = [datetime(2018, 1, 1) + timedelta(hours = i) for i in range(24 * days)]
y = np.arange(len(x))
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y)
xfmt = mdate.DateFormatter("%Y-%m-%d")
ax.xaxis.set_major_formatter(xfmt)
fig.show()
```
